```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
plt.style.use('fivethirtyeight')
```

```python
df = pd.read_csv('../../data/covid-dunia/raw/covid_19_data.csv')
df.ObservationDate = pd.to_datetime(df.ObservationDate)
```

```python
df_indo = df[df['Country/Region'] == 'Indonesia']
df_indo
```

```python
df_indo['Province/State'].unique()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=df_indo.ObservationDate, y=df_indo.Confirmed, name='Confirmed',fill='tozeroy',fillcolor='rgba(0, 0, 255, .3)'))
fig.add_trace(go.Scatter(x=df_indo.ObservationDate, y=df_indo.Deaths, name='Deaths', fill='tozeroy',fillcolor='rgba(255, 0, 0, 1)'))
fig.add_trace(go.Scatter(x=df_indo.ObservationDate, y=df_indo.Recovered, name='Recovered', fill='tozeroy',fillcolor='rgba(0, 255, 0, .3)'))

fig.update_layout(title='COVID-19 in Indonesia', xaxis_title='Date', yaxis_title='Number of Cases',)
```

```python
fig = px.histogram(df_indo, x="ObservationDate", y=["Confirmed", "Recovered","Deaths"], title="Covid-19 in Indonesia",barmode='overlay',color_discrete_sequence=['Blue','Green','Red'])
fig.show()
```

```python
df_indo['Percentage_Deaths'] = df_indo.Deaths/df_indo.Confirmed*100
df_indo['Percentage_Recovered'] = df_indo.Recovered/df_indo.Confirmed*100
```

```python
df_indo
```

```python
fig = px.histogram(df_indo, x="ObservationDate", y=["Percentage_Deaths", "Percentage_Recovered"], title="Percentage Deaths vs Recovered",color_discrete_sequence=['Red','Green'],text_auto=True,histfunc='avg')
fig.update_traces(textposition='inside',textfont_size=10)
fig.update_layout(
    title='Percentage Recovered vs Deaths',
    # xaxis_tickfont_size=14,
    uniformtext_minsize=10, uniformtext_mode=False,
    yaxis=dict(
        title='Percentage',
        titlefont_size=16,
        tickfont_size=14,
    ),
    legend=dict(
        x=0,
        y=1.0,
        bgcolor='rgba(255, 255, 255, 0)',
        bordercolor='rgba(255, 255, 255, 0)'
    ),
    barmode='stack',
    bargap=0.15, # gap between bars of adjacent location coordinates.
    bargroupgap=0.1 # gap between bars of the same location coordinate.
)
fig.show()
```

```python
# load the data
df_mobility = pd.read_csv('2020_ID_Region_Mobility_Report.csv')
print(df_mobility.shape)
df_mobility.head()
```

```python
df_mobility.isna().sum()
```

```python
df_mobility.date = pd.to_datetime(df_mobility.date)
df_mobility.columns
```

```python
# function for plot
def plot_this(data,y):
    tes = data[['date',y]]
    tes = tes.groupby('date').mean()
    tes['naik'] = np.where(tes[y]>0,tes[y],0)
    tes['turun'] = np.where(tes[y]<0,tes[y],0)
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=tes.index, y=tes.turun*10,fill='tozeroy',fillcolor='rgba(255, 0, 0,0.7)',line = dict(color='rgba(255,0,0,0.7)')))
    fig.add_trace(go.Scatter(x=tes.index, y=tes.naik*10,fill='tozeroy',fillcolor='rgba(33, 168, 96,0.7)',line = dict(color='rgba(33,168,96,0.7)')))
    fig.update_layout(
        title=' '.join(y.split('_')),
        showlegend=False,
        yaxis= {'tickformat': ',.0%'}
    )
    fig.show()
```

```python
plot_this(df_mobility,'retail_and_recreation_percent_change_from_baseline')
```

```python
plot_this(df_mobility,'grocery_and_pharmacy_percent_change_from_baseline')
```

```python
plot_this(df_mobility,'parks_percent_change_from_baseline')
```

```python
plot_this(df_mobility,'transit_stations_percent_change_from_baseline')
```

```python
plot_this(df_mobility,'workplaces_percent_change_from_baseline')
```

```python
plot_this(df_mobility,'residential_percent_change_from_baseline')
```
